```python
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModel

# Read the DataFrame
df = pd.read_csv('/content/Hate_Speech_with_bias - Sheet1.csv')

# Convert columns to string to avoid any issues
df['Male Centric Sentence'] = df['Male Centric Sentence'].astype(str)
df['Female Centric Sentence'] = df['Female Centric Sentence'].astype(str)

# Get male and female centric sentences
male_centric_sentences = df['Male Centric Sentence'].tolist()
female_centric_sentences = df['Female Centric Sentence'].tolist()

# Load the tokenizer and model for Bert
tokenizer_bert = AutoTokenizer.from_pretrained('FredZhang7/one-for-all-toxicity-v3')
model_bert = AutoModel.from_pretrained('FredZhang7/one-for-all-toxicity-v3', output_attentions=True)

# Function to get attention scores from the [CLS] token to each token
def get_cls_attention(attentions):
    # Get the last layer's attention
    last_layer_attention = attentions[-1]  # (num_layers, batch_size, num_heads, seq_len, seq_len)
    # Get the average attention from all heads for the [CLS] token (position 0) to each token
    cls_attention = last_layer_attention.mean(dim=1)[:, 0, :]  # (batch_size, seq_len)
    return cls_attention.squeeze()

# Function to find the highest attention value word for both sentences, excluding [CLS] and [SEP] tokens
def get_highest_attention_token(tokens, attention_scores):
    filtered_tokens_scores = [(token, score.item()) for token, score in zip(tokens, attention_scores) if token not in ['[CLS]', '[SEP]']]
    highest_token, highest_score = max(filtered_tokens_scores, key=lambda x: x[1])
    return highest_token, highest_score

# Function to find the longest common subsequence
def longest_common_subsequence(str1, str2):
    m, n = len(str1), len(str2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    lcs = ""
    while m > 0 and n > 0:
        if str1[m - 1] == str2[n - 1]:
            lcs = str1[m - 1] + lcs
            m -= 1
            n -= 1
        elif dp[m - 1][n] > dp[m][n - 1]:
            m -= 1
        else:
            n -= 1
    return lcs

# Function to check if the token is a gender term using LCS
def check_gender_term(token, masculine_terms, feminine_terms):
    for term in masculine_terms:
        if longest_common_subsequence(token, term) == term:
            return 'MA'
    for term in feminine_terms:
        if longest_common_subsequence(token, term) == term:
            return 'FA'
    return 'NaN'

# Example lists of masculine and feminine terms
df_gender_terms = pd.read_csv('/content/Copy of Gendered Terms(Bangla) - Sheet1.csv')
Masculine_term = df_gender_terms['Masculine'].tolist()
Feminine_term = df_gender_terms['Feminine'].tolist()

# Initialize lists to store attention scores and highest tokens
attention_labels = []

# Tokenize the sentences using Bert tokenizer with padding and truncation
for male_sentence, female_sentence in zip(male_centric_sentences, female_centric_sentences):
    # Tokenize male and female sentences separately
    male_tokenized = tokenizer_bert(male_sentence, return_tensors='pt', padding=True, truncation=True, max_length=128)
    female_tokenized = tokenizer_bert(female_sentence, return_tensors='pt', padding=True, truncation=True, max_length=128)

    # Get model outputs (including attentions) for Bert model
    with torch.no_grad():
        outputs_male_bert = model_bert(**male_tokenized)
        outputs_female_bert = model_bert(**female_tokenized)

    # Get attention scores from the [CLS] token to each token
    attention_scores_male = get_cls_attention(outputs_male_bert.attentions)
    attention_scores_female = get_cls_attention(outputs_female_bert.attentions)

    # Get the tokens for Bert model
    tokens_male_bert = tokenizer_bert.convert_ids_to_tokens(male_tokenized['input_ids'].squeeze().tolist())
    tokens_female_bert = tokenizer_bert.convert_ids_to_tokens(female_tokenized['input_ids'].squeeze().tolist())

    # Find the highest attention value word for both sentences, excluding [CLS] and [SEP] tokens
    highest_token_male, highest_score_male = get_highest_attention_token(tokens_male_bert, attention_scores_male)
    highest_token_female, highest_score_female = get_highest_attention_token(tokens_female_bert, attention_scores_female)

    # Append attention labels based on highest attention tokens
    attention_label_male = check_gender_term(highest_token_male, Masculine_term, Feminine_term)
    attention_label_female = check_gender_term(highest_token_female, Masculine_term, Feminine_term)
    if attention_label_male != 'NaN':
        attention_labels.append('MA')
    elif attention_label_female != 'NaN':
        attention_labels.append('FA')
    else:
        attention_labels.append('NaN')

# Create the final DataFrame with the new columns
df_final = pd.DataFrame({
    'male_centric_sentence': male_centric_sentences,
    'female_centric_sentence': female_centric_sentences,
    'Attention': attention_labels
})

# Calculate and print counts
total_rows = len(df_final)
total_ma = attention_labels.count('MA')
total_fa = attention_labels.count('FA')
total_blank = attention_labels.count('NaN')

print(f"Total Bias: {total_rows}")
print(f"Total 'Male Attention': {total_ma}")
print(f"Total 'Female Attention': {total_fa}")
print(f"Total Other Attention Rows: {total_blank}")

print("\nUpdated DataFrame with Attention labels:")
print(df_final.head())
```
