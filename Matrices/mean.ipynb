```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load your dataframe from the CSV file
df = pd.read_csv('/content/Sarcasm_Final_&_Prediction_results.csv')

# Load the Bangla BERT model and tokenizer
model_name = 'raquiba/sarcasm-detection-BanglaSARC'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Function to get softmax scores
def get_softmax_scores(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        outputs = model(**inputs)
    softmax_scores = torch.softmax(outputs.logits, dim=1)
    return softmax_scores.numpy().squeeze()  # Remove the extra dimension

# Process 'Original Sentence' column
df['softmax_scores_Original'] = df['Original Sentence'].apply(get_softmax_scores)

# Assuming the model outputs three classes: negative, neutral, positive
softmax_columns_original = ['class_0_Original', 'class_1_Original', 'class_2_Original']

# Split the softmax scores into separate columns for Original Sentence
softmax_df_original = pd.DataFrame(df['softmax_scores_Original'].tolist(), columns=softmax_columns_original)

# Determine the class with the highest softmax score for Original Sentence
df['predicted_class_Original'] = softmax_df_original.values.argmax(axis=1)

# Determine the highest softmax score for Original Sentence
df['highest_softmax_score_Original'] = softmax_df_original.max(axis=1)

# Concatenate the softmax scores dataframe for Original Sentence
df = pd.concat([df, softmax_df_original], axis=1)

# Process 'Converted Sentence' column
df['softmax_scores_Converted'] = df['Converted Sentence'].apply(get_softmax_scores)

# Adjust the softmax columns for Converted Sentence
softmax_columns_converted = ['class_0_Converted', 'class_1_Converted', 'class_2_Converted']

# Split the softmax scores into separate columns for Converted Sentence
softmax_df_converted = pd.DataFrame(df['softmax_scores_Converted'].tolist(), columns=softmax_columns_converted)

# Determine the class with the highest softmax score for Converted Sentence
df['predicted_class_Converted'] = softmax_df_converted.values.argmax(axis=1)

# Determine the highest softmax score for Converted Sentence
df['highest_softmax_score_Converted'] = softmax_df_converted.max(axis=1)

# Concatenate the softmax scores dataframe for Converted Sentence
df = pd.concat([df, softmax_df_converted], axis=1)

# Filter dataframe for class 0 and class 1
class_0_1_df = df[(df['predicted_class_Converted'] == 0) | (df['predicted_class_Converted'] == 1)]

# Print the resulting dataframe for class 0 and class 1
print(class_0_1_df[['Original Sentence', 'Converted Sentence', 'predicted_class_Original', 'highest_softmax_score_Original', 'predicted_class_Converted', 'highest_softmax_score_Converted']].head())  # Print only relevant columns and the first few rows for brevity
```

```python
df['row_value_original'] = df['highest_softmax_score_Original'].apply(lambda x: abs(x - 0.5))
df['row_value_converted'] = df['highest_softmax_score_Converted'].apply(lambda x: abs(x - 0.5))
# Calculate the sum of these absolute differences
sum_abs_diff_origi = df['row_value_original'].sum()
sum_abs_diff_conver = df['row_value_converted'].sum()

# Calculate the total number of rows
total_rows = len(df)

# Compute the mean bias confidence
mean_bias_confidence_Original = (1 / total_rows) * sum_abs_diff_origi
mean_bias_confidence_Converted = (1 / total_rows) * sum_abs_diff_conver

print("Original Sentence Mean Bias Confidence:", mean_bias_confidence_Original)
print("Converted Sentence Mean Bias Confidence:", mean_bias_confidence_Converted)
# print(df[['predicted_class_softmax', 'row_value']].head())
```
