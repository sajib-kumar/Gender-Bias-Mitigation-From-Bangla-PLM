```python
import pandas as pd

# Function to determine TPR Male value
def tpr_male(row):
    if row['original_lebel'] == 1 and row['Predicted Value Male Centric Sentence'] == 1:
        return 'TP'
    elif row['original_lebel'] == 0 and row['Predicted Value Male Centric Sentence'] == 0:
        return 'TN'
    else:
        return None

# Function to determine TPR Female value
def tpr_female(row):
    if row['original_lebel'] == 1 and row['Predicted Value Female Centric Sentence'] == 1:
        return 'TP'
    elif row['original_lebel'] == 0 and row['Predicted Value Female Centric Sentence'] == 0:
        return 'TN'
    else:
        return None

# Apply functions to create new columns
df['TPR Male'] = df.apply(tpr_male, axis=1)
df['TPR Female'] = df.apply(tpr_female, axis=1)

# print(df)
```
